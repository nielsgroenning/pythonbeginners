```python
import swat
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
#Connectio to CAS and establish a session named "s"
s = swat.CAS('URL','PORT','USERNAME','PASS', protocol='http')
```

```python
# Assign the variable name dm to the new CASTable object
cas_pokemon = s.CASTable("cas_pokemon", replace=True)

#The s.upload_file() method uploads a data file as-is to CAS and invokes the table.loadtable action in the background to parse it.
s.upload_file('https://raw.githubusercontent.com/nielsgroenning/pythonbeginners/master/Pokemon_data.csv',casout=cas_pokemon)
```

```python
cas_pokemon.head()
```

```python
#Display Summary Statistics 
cas_pokemon.groupby('Type 1').summary(subset=['Mean','Min','Max'])
```

```python
#Drop the 'Generation', '#' and 'Legendary' Columns
cas_pokemon.drop(['Generation','#', 'Legendary'], axis=1, inplace=True)

#Display new column
cas_pokemon.head()
```

```python
#Display Pokemons Type (1) using matplotlib
cas_pokemon.plot.scatter(x='Total', y='HP', figsize=(7,7))
```

```python
#Extract Pokemon Types to new variable
Pokemon_Type = cas_pokemon['Type 1'].value_counts()

#This method fetches the data from the CAS table and calls the pandas.DataFrame.plot.pie() method to do the rendering.
Pokemon_Type.plot.pie(y='Type 1', figsize=(7, 7), autopct='%1.1f%%')
```
